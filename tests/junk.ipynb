```python
import lmdb
import struct
env = lmdb.Environment("TEMP_DB", readonly=True, max_dbs=10)

def encode_int(i):
    return struct.pack('i', i)

def decode_int(i_bytes):
    return struct.unpack('i', i_bytes)[0]
def decode_int64(i_bytes):
    return struct.unpack('q', i_bytes)[0]

with env.begin() as txn:
    subdb = env.open_db(b"heavy_dimers", txn, create=False)
    cursor = lmdb.Cursor(subdb, txn)
    for key, value in cursor:
        print(f"Key: {decode_int(key)}, Value: {decode_int64(value)}")
        print(''.join(f"{byte:08b}" for byte in value))
```
