```python
import gzip
from antpack import SingleChainAnnotator as SCA
sca = SCA()
```

```python
with gzip.open("vj_gene_testing.csv.gz", "rt") as fhandle:
    lines = fhandle.readlines()

for i, line in enumerate(lines[1:]):
    seq = line.split(",")[0]
    annot, pid, chain, err = sca.analyze_seq(seq)
    updated_line = lines[i+1].strip()
    updated_line += "," + chain + '\n'
    lines[i+1] = updated_line
```

```python
lines[0] = lines[0].strip() + "," + "chain" + "\n"
```

```python
with gzip.open("vj_gene_testing.csv.gz", "wt") as fhandle:
    for line in lines:
        fhandle.write(line)
```
